```python
import pandas as pd
# Example: Replace this with the raw URL of your GitHub file
github_raw_url = "https://raw.githubusercontent.com/Nithi711/ADALL_github/refs/heads/main/laptop_prices_2024_sgd_TL.csv"
try:
  df = pd.read_csv(github_raw_url)
  print("Successfully loaded data from GitHub!")
  display(df.head())   # Display give the data more strutured. Print give you the top 5 in text format.

  #import plt library
  import matplotlib.pyplot as plt

  #plot each of df col graph
  df.plot(kind='bar')
  plt.show()

  #plot histogram for every numerical column of df col
  df.hist(bins=50, figsize=(12,8))
  plt.show()

except Exception as e:
  print(f"Error loading data: {e}")
  print("Please ensure the URL is correct and the file format is compatible with `pd.read_csv`.")
```
